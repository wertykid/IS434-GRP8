```python
import json
import pandas as pd

# Opening JSON file
f = open('shimanobikes_combined.json')
 
# returns JSON object as
# a dictionary
data = json.load(f)

#which dataset are you using? e.g. unsprung ,Reevo ...
postOwner = "shimanomtb"

file_name = 'ShimanoFilteredHashtags.json'
#save it into a list so that its easier for pandas
initialResult = []
# # Iterating through the json
# # list
# Here, I want to put it into this format so that we can translate it into a networkX graph
# Essentially we are saying this post by __post owner__ used this hashtag
# the hashtag could be tagging another company or bike .
# We can find out if the companies tag each other? or is it one directional?

for post in data:
    hashes = post["hashtags"]
    for hashtag in hashes:
        if hashtag != postOwner:
            forDataFrame = {
                "postBy" : "Shimanomtb",
                "postHashtag" : hashtag
            }
            initialResult.append(forDataFrame)


result = json.dumps(initialResult)

with open(file_name, 'a+') as filehandle:
    filehandle.write(result)
# print(initialResult)
 
df1 = pd.json_normalize(initialResult)





df1
```
